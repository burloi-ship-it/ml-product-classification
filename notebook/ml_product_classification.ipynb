```python
import pandas as pd

# load dataset from GitHub
url = "https://raw.githubusercontent.com/burloi-ship-it/ml-product-classification/refs/heads/main/data/products.csv"

# load data into a dataframe
df = pd.read_csv(url)

# first 5 rows
print(f"\nFirst 5 rows: \n")
display(df.head(5))

# number of rows in dataframe
print(f"\nRows in dataframe: \n")
print(df.shape[0])

# number of missing values
print(f"\nNumber of missing values: \n")
print(df.isna().sum())

# column data types and non null counts
print("\nColumn data types: ")
print(df.info())
```

```python
# rename columns for clear view and better processing
print(f"Renaming columns in dataset...")
df = df.rename(columns={'product ID': 'product_id', 'Product Title': 'product_title', 'Merchant ID': 'merchant_id', ' Category Label': 'category_label', '_Product Code': 'product_code', 'Number_of_Views': 'number_of_views', 'Merchant Rating': 'merchant_rating', ' Listing Date  ': 'listing_date'})
print(f"Done!")

# stergerea valorilor NaN
print(f"\nDropping missing values in dataset...")
df = df.dropna()
print(f"Done!")
# verificarea numarului de valori nule din coloane
print("\nCurrent number of missing values: ")
print(df.isna().sum())

# standardizarea textului coloanelor 'product_title' si 'category_label'
print(f"\nStandardizing 'product_title' and 'category_label' columns to lowercase text...")
df['product_title'] = df['product_title'].str.lower()
df['category_label'] = df['category_label'].str.lower()
print(f"Done!")

# conversia coloanei 'listing_date' in format datetime
print(f"\nConverting 'listing_date' column in datetime format...")
df['listing_date'] = pd.to_datetime(df['listing_date'], format="%m/%d/%Y", errors="coerce")
print(f"Done!")

# conversia coloanei 'category_label' in tip category
print(f"\nConverting 'category_label' column in object type...")
df['category_label'] = df['category_label'].astype('category')
print(f"Done!")

# conversia coloanei 'number_of_views' in tip Int64
print(f"\nConverting 'number_of_views' column in Int64 type...")
df['number_of_views'] = df['number_of_views'].astype('Int64')
print(f"Done!")

# listarea coloanelor actuale
print("\nListing current columns in dataset: ")
print(*df.columns, sep='\n')

# listarea tipului de date ale coloanelor dupa conversie
print(f"\nNew available columns data types: \n")
print(df.info())
```

```python
# dropping irrelevant columns
print(f"\nDropping irrelevant columns for model training and prediction...")
df = df.drop(columns=['product_id','product_code', 'number_of_views', 'merchant_rating', 'listing_date'])
print(f"Done!")

# listing actual dataframe columns
print(f"\nCurrent dataframe available columns: ")
print(*df.columns, sep='\n')
```

```python
# function to extract brand and storage
import re
def extract_storage(title):
  match = re.search(r'(\d+)\s*gb', title.lower())
  return int(match.group(1)) if match else None

# extract brand
def extract_brand(title, brand_list):
  title_lower = title.lower()
  for brand in brand_list:
    if brand in title_lower:
      return brand
  return 'Unknown'

# color extraction
def extract_color(title, color_list):
  title_lower = title.lower()
  for color in color_list:
    if color in title_lower:
      return color
  return 'Unknown'

# storage single argument function , apply directly
df['storage_gb'] = df['product_title'].apply(extract_storage)

# brand - use lambda for brand_list argument
brand_list = ['apple', 'samsung', 'huawei', 'xiaomi', 'nokia', 'sony', 'google', 'oneplus']
df['brand'] = df['product_title'].apply(lambda x: extract_brand(x, brand_list))

# color extraction
color_list = ['silver', 'black', 'white', 'gold', 'spacegrau', 'space grey', 'blue', 'red']
df['color'] = df['product_title'].apply(lambda x: extract_color(x, color_list))

df['category_label'] = (
    df['category_label']
    .str.strip()
    .str.lower()
)

display(df)
```

```python
# split data for train and test
from sklearn.model_selection import train_test_split

# Training models on data
# Features and labels
x = df[['product_title', 'merchant_id', 'storage_gb', 'brand', 'color']]
y = df['category_label']

# Splitting data in train data 80% and test data 20%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# vectorization
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

numeric_pipeline = Pipeline([('impute', SimpleImputer(strategy='median'))])

preprocessor = ColumnTransformer([
    ('title', TfidfVectorizer(analyzer='char_wb', ngram_range=(2, 4)), 'product_title'), # spacegrau vs space grey
    ('merchant', OneHotEncoder(handle_unknown='ignore'), ['merchant_id']),
    ('storage', numeric_pipeline, ['storage_gb']),
    ('brand', OneHotEncoder(handle_unknown='ignore'), ['brand']),
    ('color', OneHotEncoder(handle_unknown='ignore'), ['color'])
])
```

```python
# model testing
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

# list of classifiers
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000, class_weight='balanced', solver='saga'),
    "Naive Bayes": MultinomialNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": LinearSVC(max_iter=5000)
}

# train and test models
for name, model in models.items():
  print(f"\n{name}")
  pipeline = Pipeline([
      ("preprocessing", preprocessor),
      ("classifier", model)
  ])

  pipeline.fit(x_train, y_train)
  y_prediction = pipeline.predict(x_test)
  print(classification_report(y_test, y_prediction, zero_division=0))
```
