```python
import pandas as pd

# load dataset from GitHub
url = "https://raw.githubusercontent.com/burloi-ship-it/ml-product-classification/refs/heads/main/data/products.csv"

# load data into a dataframe
df = pd.read_csv(url)

# first 5 rows
print(f"\nFirst 5 rows: \n")
display(df.head(5))

# number of rows in dataframe
print(f"\nRows in dataframe: \n")
print(df.shape[0])

# number of missing values
print(f"\nNumber of missing values: \n")
print(df.isna().sum())

# column data types and non null counts
print("\nColumn data types: ")
print(df.info())
```

```python
# rename columns for clear view and better processing
print(f"Renaming columns in dataset...")
df = df.rename(columns={'product ID': 'product_id', 'Product Title': 'product_title', 'Merchant ID': 'merchant_id', ' Category Label': 'category_label', '_Product Code': 'product_code', 'Number_of_Views': 'number_of_views', 'Merchant Rating': 'merchant_rating', ' Listing Date  ': 'listing_date'})
print(f"Done!")

# stergerea valorilor NaN
print(f"\nDropping missing values in dataset...")
df = df.dropna()
print(f"Done!")
# verificarea numarului de valori nule din coloane
print("\nCurrent number of missing values: ")
print(df.isna().sum())

# standardizarea textului coloanelor 'product_title' si 'category_label'
print(f"\nStandardizing 'product_title' and 'category_label' columns to lowercase text...")
df['product_title'] = df['product_title'].str.lower()
df['category_label'] = df['category_label'].str.lower()
print(f"Done!")

# conversia coloanei 'listing_date' in format datetime
print(f"\nConverting 'listing_date' column in datetime format...")
df['listing_date'] = pd.to_datetime(df['listing_date'], format="%m/%d/%Y", errors="coerce")
print(f"Done!")

# conversia coloanei 'category_label' in tip category
print(f"\nConverting 'category_label' column in object type...")
df['category_label'] = df['category_label'].astype('category')
print(f"Done!")

# conversia coloanei 'number_of_views' in tip Int64
print(f"\nConverting 'number_of_views' column in Int64 type...")
df['number_of_views'] = df['number_of_views'].astype('Int64')
print(f"Done!")

# listarea coloanelor actuale
print("\nListing current columns in dataset: ")
print(*df.columns, sep='\n')

# listarea tipului de date ale coloanelor dupa conversie
print(f"\nNew available columns data types: \n")
print(df.info())
```

```python
# dropping irrelevant columns
print(f"\nDropping irrelevant columns for model training and prediction...")
df = df.drop(columns=['product_id', 'merchant_id', 'product_code', 'number_of_views', 'merchant_rating', 'listing_date'])
print(f"Done!")

# listing actual dataframe columns
print(f"\nCurrent dataframe available columns: ")
print(*df.columns, sep='\n')
```
